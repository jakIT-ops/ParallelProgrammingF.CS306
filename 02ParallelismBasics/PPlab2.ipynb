#**Параллелчлалын үндэс**

##**Параллел алгоритмын шинжилэгээ**


<font color="red">Хурдсалт(Speed up)</font> : Хурдсалт(S) нь p ширхэг процессор дээр ажиллах хугацаа (T(p)) болон нэг процессор дээр ажиллах хугацаа (T(1)) - уудаа харьцаа юм.
```
S = T(1) / T(p)
```
<font color="yellow">Үр ашиг ба өртөг (Effiency and cost)</font> : S - ийг p-д хувааж үр ашиг (Efficiency) E-г олно
```
E = S / p
```
Өртөг(Cost) нь төстэй боловч ажиллах хугацаа T(p) (хурдсалтын оронд)-г цөмийн тоо p-ээр үржүүлдэг.
```
C = T(p) x p
```
<font color="green">Чадамж (Scalability)</font> : Ихэнхдээ тодорхой тооны процессор эсвэл цөмүүдийн үр ашгийг хэмжихийн оронд өөр өөр тооны хувьд шинждэг. P = 1,2,4,8,16...

**Strong scalability:** Процессоруудын тоог нэмэгдүүлж шинжлэх

**Weak scalability:** Процессоруудын тоо болон оролтын өгөгдлийн хэмжээг өсгөж шинжилнэ.

<font color="blue">Computation-to-communication харьцаа</font> : Параллел хэрэгжүүлэлтийн боломжит чадамжид нөлөөлөх чухал хэмжигдэхүүн юм. 

Тооцоололд зарцуулсан хугацааг процессоруудын хооронд мэдээлэл солилцооны хугацаагаар хувааж тооцоолно. 

Өндөр байх тусам хурдсалт, үр ашиг нэмэгдэнэ.

##**Параллел программын зохиомж**

<font color="red">Partitioning</font> : Өгөгдсөн асуудлыг дэд хэсгүүдэд задрах шаардлагатай. Жишээ нь өгөгдлийн, даалгаварын болон загварын параллелизм гэсэн схемүүдээс сонгох хуваалт хийнэ.

<font color="yellow">Communication</font> : Сонгосон хуваалтын схем нь шаардлагатай мэдээлэл солилцооны хэмжээ, төрлийн тодорхойлно. 

<font color="green">Synchronization</font> : Тохиромжтой байдлаар хамтран ажиллахын тулд thread эсвэл процессыг синхрончлох шаардлагатай байж болно.

<font color="blue">Load Balancing</font> : Ачааллыг тэнцвэржүүлэх, сул зогсолтыг багасгахын тулд ажлыг thread болон процесст тэнцүү хуваах шаардлагатай.

#**Parallel Random Access Machine(PRAM)**

**Дундын санах ойн хийсвэр машин юм.**
*   Процессорууд хоорондоо дундын санах ойд уншиж, бичиж байдлаар харилцана.

<font color="green">Exclusice Read Exclusive Write(EREW): </font> Ямар ч циклийн явцад дундын санах ойн нэг үүрэнд хоёр процессор уншиж, бичихийг зөвшөөрдөггүй.

<font color="red">Concurrent Read Exclusive Write(CREW): </font> Зарим процессорууд дундын санах ойн нэг үүрнээс зэрэг унших боломжтой. Гэсэн ч өөр процессорууд дундын санах ойн нэг үүрэнд бичихийг зөвшөөрдөггүй.

<font color="blue">Concurrent Read Concurrent Write(CRCW): </font> Дундын санах ойн үүрэнд зэрэгцээгээр уншиж бичдэг. Зэрэг бичилт хийх тохиолдолд (Уралдааны нөхцөлд хүргэх) аль утгыг нь хадаглахыг тодорхой зааж өгнө.

**Contradictory requirements**

Сүлжээний зохиомж нь **зөрчилдөөнтөй шаардлагуудын** хооронд уялдааг хангах эрмэлзэлтэй.

* **Constant degree:** Сүлжээний degree тогтмол байх ёстой Жнь Сүлжээг өргөтгөхөд нь холболтыг нэмэх шаардлагагүй.

* **Low diameter:** Аливаа хоёр процессорын холболтыг үр ашигтай байлгах үүднээс диаметрийг хамгийн бага байлгах хэрэгтэй.

* **High bisection-width:** bisection-width нь сүлжээний боломжит чадварыг илэрхийлэх бөгөөд энэ нь дотоод bandwith-д нөлөөлнө.

##Foster's Parallel Algorithm Design Method

*   **Partitioning:** асуудлыг параллел ажилуулж болох олон жижиг даалгавруудад (fine-grained) задлана. 
*   **Communication:** даалгавруудын хоорондох шаардлагатай мэдээлэл солилцоог тодорхойлох
*   **Agglomeration:** Өгөгдлийн байршлыг сайжруулан мэдээлэл солилцоог багасгах зорилгоор тодорхойлсон даалгаваруудыг том даалгаварт нэгтгэнэ.
*   **Mapping:** Мэдэлэл солилцоог багасгах, давхардалгүй байдлыг хангах болон ачааллыг тэнцвэржүүлэх зорилгоор бөөгнөрлийг процесс / thread-үүдэд хуваарилна.

##Жишээ нь: Jacobi Iteration

*   Матриц дахь утга бүрийг дөрвөн хөршийнх дундаж утгаар солино
*   Хүрээний утгууд нь тогтмол байна

**Конвергенци болох хүртлэх давталтын алхам бүрт өгөгдсөн 2D матрицын бүх цэгийн утгыг хөршүүдийнх нь дундажаар солино.**
```
while (not converged) {
  for (int i = 1; i < rows-1; i++)
    for (int j = 1; j<cols-1; j++)
      buff[i*cols+j] = 0.25f*(data[(i+1)*cols+1] + data[i*cols+j-1] + data[i*cols+j+1] + data[(i-1)*cols+j]);
      
  memcpy(data,buff, rows*cols*sizeof(float));    
}
```
